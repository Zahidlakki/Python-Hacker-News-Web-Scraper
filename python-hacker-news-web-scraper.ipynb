```python
# Web Scraper: Hacker News Headlines
# Author: Zahid Ali
# Description: This script scrapes news headlines from Hacker News
# and saves them into a text file automatically.

import requests
from bs4 import BeautifulSoup

# Website URL
url = "https://news.ycombinator.com/"

# Send request to website
response = requests.get(url)

# Convert HTML content into text
html = response.text

# Parse HTML using BeautifulSoup
soup = BeautifulSoup(html, "html.parser")

# Find all headline elements
headlines = soup.find_all("span", class_="titleline")

# Save headlines to a file
with open("headlines.txt", "w", encoding="utf-8") as file:
    for headline in headlines:
        file.write(headline.text + "\n")

print("✅ Headlines scraped and saved successfully!")
```
